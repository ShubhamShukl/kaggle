```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_excel('/kaggle/input/keywords1/Keyword and hashtag1.xlsx')

# Set the style for the plot
sns.set(style="whitegrid")

```

```python
df
```

```python
df = df.drop(index=2)
df.reset_index(drop=True, inplace=True)
```

```python
# Create a new figure
plt.figure(figsize=(12, 8))

# Convert the dataframe to long format for easier plotting
df_long = df.melt(id_vars=['Keyword', 'Hashtag/Term'], 
                  value_vars=['Original Count', 'Final Count'], 
                  var_name='Count Type', value_name='Count')

# Create the grouped bar plot
sns.barplot(x='Keyword', y='Count', hue='Count Type', data=df_long, palette='coolwarm', dodge=True)

# Add title and labels
plt.title('Original vs Final Tweet Counts by Keyword (Grouped by Hashtag/Term)', fontsize=16)
plt.xlabel('Keyword', fontsize=14)
plt.ylabel('Tweet Count', fontsize=14)

# Rotate the x labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a legend to differentiate "Hashtag" and "Term"
plt.legend(title='Count Type')

# Save the plot as JPEG and PNG
plt.savefig('tweet_counts_plot.jpeg', format='jpeg', dpi=400)
plt.savefig('tweet_counts_plot.png', format='png', dpi=400)

# Display the plot
plt.tight_layout()
plt.show()
```
